```python
import os
import geopandas as gpd
import rasterio
from rasterio.plot import show
import pandas as pd

raster_dir = r'D:\Landslide\data\processed\gyeongnam\DEM_outputs'

print("ğŸ—ºï¸ DEM ë˜ìŠ¤í„° íŒŒì¼ ë¶„ì„")
print("="*80)

# TIF íŒŒì¼ ëª©ë¡ ìˆ˜ì§‘
tif_files = [f for f in os.listdir(raster_dir) if f.endswith('.tif')]
print(f"ë°œê²¬ëœ TIF íŒŒì¼: {len(tif_files)}ê°œ\n")

if not tif_files:
    print(f"âŒ TIF íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {raster_dir}")
else:
    # ë˜ìŠ¤í„° ì •ë³´ ìˆ˜ì§‘
    raster_info = []
    
    for idx, filename in enumerate(sorted(tif_files)):
        filepath = os.path.join(raster_dir, filename)
        
        try:
            with rasterio.open(filepath) as src:
                info = {
                    'íŒŒì¼ëª…': filename,
                    'í¬ê¸° (í–‰xì—´)': f"{src.height} x {src.width}",
                    'ë°´ë“œ ìˆ˜': src.count,
                    'ì¢Œí‘œê³„ (CRS)': str(src.crs),
                    'í”½ì…€ í¬ê¸°': f"{abs(src.transform[0]):.1f}m x {abs(src.transform[4]):.1f}m",
                    'ë²”ìœ„ (minx, miny, maxx, maxy)': f"({src.bounds.left:.0f}, {src.bounds.bottom:.0f}, {src.bounds.right:.0f}, {src.bounds.top:.0f})",
                    'ë°ì´í„° íƒ€ì…': str(src.dtypes[0]),
                    'íŒŒì¼ í¬ê¸°': f"{os.path.getsize(filepath) / 1024**2:.1f} MB"
                }
                raster_info.append(info)
                
                print(f"ğŸ“‹ [{idx+1}] {filename}")
                print(f"    í¬ê¸°: {info['í¬ê¸° (í–‰xì—´)']} ({info['ë°´ë“œ ìˆ˜']}ë°´ë“œ)")
                print(f"    ì¢Œí‘œê³„: {info['ì¢Œí‘œê³„ (CRS)']}")
                print(f"    í”½ì…€ í¬ê¸°: {info['í”½ì…€ í¬ê¸°']}")
                print(f"    ë²”ìœ„: {info['ë²”ìœ„ (minx, miny, maxx, maxy)']}")
                print(f"    ë°ì´í„° íƒ€ì…: {info['ë°ì´í„° íƒ€ì…']} | íŒŒì¼ í¬ê¸°: {info['íŒŒì¼ í¬ê¸°']}")
                print()
                
        except Exception as e:
            print(f"âŒ [{idx+1}] {filename} - ì½ê¸° ì‹¤íŒ¨: {e}")
            print()
    
    # DataFrameìœ¼ë¡œ ì •ë¦¬
    if raster_info:
        df = pd.DataFrame(raster_info)
        
        print("="*80)
        print("ğŸ“Š ìš”ì•½ ì •ë³´")
        print("="*80)
        
        # ì¢Œí‘œê³„ í†µê³„
        crs_counts = df['ì¢Œí‘œê³„ (CRS)'].value_counts()
        print("\nğŸŒ ì¢Œí‘œê³„ ë¶„í¬:")
        for crs, count in crs_counts.items():
            print(f"  - {crs}: {count}ê°œ íŒŒì¼")
        
        # í¬ê¸° í†µê³„
        unique_sizes = df['í¬ê¸° (í–‰xì—´)'].value_counts()
        print("\nğŸ“ ë˜ìŠ¤í„° í¬ê¸° ë¶„í¬:")
        for size, count in unique_sizes.items():
            print(f"  - {size}: {count}ê°œ íŒŒì¼")
        
        # í”½ì…€ í¬ê¸° í†µê³„
        unique_pixel_sizes = df['í”½ì…€ í¬ê¸°'].value_counts()
        print("\nğŸ” í”½ì…€ í•´ìƒë„ ë¶„í¬:")
        for pixel_size, count in unique_pixel_sizes.items():
            print(f"  - {pixel_size}: {count}ê°œ íŒŒì¼")
        
        print(f"\nğŸ“ ì´ íŒŒì¼ í¬ê¸°: {df['íŒŒì¼ í¬ê¸°'].str.replace(' MB', '').astype(float).sum():.1f} MB")
        
        # ì •ë ¬ëœ ì „ì²´ í…Œì´ë¸” í‘œì‹œ
        print("\n" + "="*80)
        print("ğŸ“‹ ì „ì²´ ë˜ìŠ¤í„° ì •ë³´ í…Œì´ë¸”")
        print("="*80)
        
        # ë” ë‚˜ì€ í‘œì‹œë¥¼ ìœ„í•´ ì—´ ìˆœì„œ ì¡°ì •
        display_columns = ['íŒŒì¼ëª…', 'í¬ê¸° (í–‰xì—´)', 'ì¢Œí‘œê³„ (CRS)', 'í”½ì…€ í¬ê¸°', 'ë°ì´í„° íƒ€ì…', 'íŒŒì¼ í¬ê¸°']
        display_df = df[display_columns]
        
        # pandas í‘œì‹œ ì˜µì…˜ ì„¤ì •
        pd.set_option('display.max_columns', None)
        pd.set_option('display.width', None)
        pd.set_option('display.max_colwidth', 50)
        
        print(display_df.to_string(index=False))
        
        print("\nğŸ’¡ ì£¼ìš” í™•ì¸ ì‚¬í•­:")
        print("  âœ… ëª¨ë“  ë˜ìŠ¤í„°ê°€ ë™ì¼í•œ ì¢Œí‘œê³„ë¥¼ ì‚¬ìš©í•˜ëŠ”ì§€ í™•ì¸")
        print("  âœ… ëª¨ë“  ë˜ìŠ¤í„°ê°€ ë™ì¼í•œ í”½ì…€ í¬ê¸°ë¥¼ ì‚¬ìš©í•˜ëŠ”ì§€ í™•ì¸")  
        print("  âœ… ë˜ìŠ¤í„° ê°„ ê³µê°„ì  ë²”ìœ„ê°€ ì¼ì¹˜í•˜ëŠ”ì§€ í™•ì¸")
        print("  âœ… ë°ì´í„° íƒ€ì…ì´ ì ì ˆí•œì§€ í™•ì¸ (ë³´í†µ float32 ë˜ëŠ” int16)")

print("\n" + "="*80)

```
