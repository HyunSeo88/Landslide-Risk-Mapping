```python
import geopandas as gpd
import matplotlib.pyplot as plt

# 파일 경로를 실제 gpkg 파일로 지정 (예시로 경로 바꿔주세요)
file_path = r"D:\Landslide\data\processed\gyeongnam\SU\slope_properties_v3.gpkg"

# GPKG 파일 불러오기
gdf = gpd.read_file(file_path)

# area_m2 컬럼 값 추출 (결측치 제거)
area_values = gdf['area_m2'].dropna()

# 히스토그램 시각화
plt.figure(figsize=(8, 5))
plt.hist(area_values, bins=100, edgecolor='black')
plt.xlabel('Area (m²)')
plt.ylabel('Number of polygons')
plt.title('Slope Polygon Area Distribution Histogram')
plt.grid(True)
plt.tight_layout()
plt.show()
```
